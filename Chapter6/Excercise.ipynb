Mở và đóng hình thái với hình ảnh thang độ xám

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.filters.rank import maximum, minimum
from skimage.morphology import disk
image = imread('../images/lena.jpg')
if image.ndim == 3:
    image = rgb2gray(image)
    image = (image * 255).astype(np.uint8)
selem = disk(5)
eroded = minimum(image, selem)
opened = maximum(eroded, selem)
dilated = maximum(image, selem)
closed = minimum (dilated, selem)
plt.figure(figsize=(15, 10))
plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image', size=15)
plt.axis('off')
plt.subplot(2,2, 2)
plt.imshow(opened, cmap='gray')
plt.title('Opened Image (Morphological Opening)', size=15)
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(closed, cmap='gray')
plt.title('Closed Image (Morphological Closing)', size=15)
plt.axis('off')
plt.tight_layout()
plt.show()
```
